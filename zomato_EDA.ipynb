```python
df_country=pd.read_excel('/content/drive/MyDrive/Country-Code.xlsx')
df_country.head(10)
```

```python
final_df=pd.merge(df,df_country,on='Country Code',how='left')
final_df.head(2)
```

```python
country_names=final_df.Country.value_counts().index
country_names
```

```python
country_val=final_df.Country.value_counts().values
country_val
```

```python
#top 3 countries with more business
plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.2f%%')
```

```python
#Observation:-zomato max transaction are in india,usa,uk
```

```python
#numerical variabes

final_df.columns
```

```python
ratings=final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
ratings
```

```python
#Observation=
#when rating btw 4.5-4.9 :-Excellent
#when rating btw 4.0-4.4 :-Very Good
#when rating btw 3.5-4.9 :-Good.....
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x=ratings['Aggregate rating'],y=ratings['Rating Count'],hue=ratings['Rating color'])
```

```python
#observation:-not rated-white
#max no of rating are between 2.5-3.9
```

```python
#count plot
sns.countplot(x=ratings['Rating color'])
```
